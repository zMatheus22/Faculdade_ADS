```python
import pandas as pd # biblioteca usada para manipulação de DataFrames

from ydata_profiling import ProfileReport # biblioteca para profiling de dados
from sklearn.model_selection import train_test_split # função para dividir o dataset em treino e teste

# importando algoritmos de ML
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
# importando o dataset
df_wine = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv", sep=';')
df_wine.describe()
```

```python
# o `to_file` permite que salvemos o relatório em um arquivo separado.
profile = ProfileReport(df_wine)
profile.to_file("relatorio_salvo_nesta_mesma_pasta.html")
```

```python
# remoção de duplicatas
df_wine = df_wine.drop_duplicates()
df_wine.shape
```

```python
# dividindo o DataFrame entre treino e teste
X_train, X_test, y_train, y_test = train_test_split(
    df_wine.drop('quality', axis=1), df_wine['quality'],
    test_size=0.33, random_state=2000)
```

```python
modelo_knn = KNeighborsRegressor(n_neighbors=2).fit(X_train, y_train)
```

```python
modelo_knn.score(X_test, y_test)
```

```python
df_test = X_test.copy()
df_test['Quality_Real'] = y_test
df_test['Quality_Predicao'] = modelo_knn.predict(X_test)
df_test
```

```python
modelo_rf = RandomForestRegressor(random_state=0).fit(X_train, y_train)
```

```python
modelo_rf.score(X_test, y_test)
```

```python
df_test['Quality_Predicao_2'] = modelo_rf.predict(X_test)
df_test
```
